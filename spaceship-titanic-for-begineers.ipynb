#### Missing Value Analysis

Every Dataset we get,is not always clean and good to use.So first we need to take a 
look at our dataset for certain checks.<br>As we saw in info method above,dataset have missing or null values in all the columns excpet PassengerId & Tranported. 
we need to check count of null values column wise. 

1. CryoSleep col has most missing values 217 and 2.49% of total entries.
2. PassengerId and Transported has no missing values at all.

```python
# count of missing values column wise in descending order
print(df_train.isnull().sum().sort_values(ascending=False))

# Just to print dotted separation line
print("-"*30)

# percentage of the missing values in the column Cryosleep(which has most missing values)
print("missing value % :",((df_train['CryoSleep'].isnull().sum())/len(df_train))*100)
```

##### Handling missing values
We need to handle missing values either by removing columns or filling missing entries.Here,max missing values is 2.49% so we need not to remove any column ,Just to fill those entries with proper filling strategy.
1. Numerical features cols should be filled with mean or median of that particular column.
2. Categorical will be filled with mode of that column.
<br>

#### Transforming Features <br>
Age, RoomService, FoodCourt, ShoppingMall, Spa, VRDeck are the features,which has numerical values.so we need to fill with either mean or median.Here I am filling those with median of respective columns. Applying on train and test data both.

```python
l = ['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
for i in l:
    df_train[i].fillna(df_train[i].median(),inplace=True)
    df_test[i].fillna(df_test[i].median(),inplace=True)
```

HomePlanet, CryoSleep, Destination, VIP are the features which have not numerical values so we can't use mean or median to fill as these can be calculated for numerical values only.So here we are using mode(most frequent value of respective col) to fill missing values.

```python
#filling categorical features with mode of respective columns
m = ['HomePlanet','CryoSleep','Destination','VIP']
for i in m:
    df_train[i].fillna(df_train[i].mode()[0],inplace=True)
    df_test[i].fillna(df_test[i].mode()[0],inplace=True)
```

```python
#Removing transported column from training dataset and storing it in to y.
y = df_train['Transported']
df_train.drop('Transported',axis=1,inplace=True)
```

We have Cabin and Name features in our dataset but These two features are not very relevant for training purpose,so we are removing these two features from our training and test data.

```python
# Drop Cabin and Name column from train data
df_train.drop(columns=['Cabin','Name'],axis=1,inplace=True)

# Drop Cabin and Name column from test data
df_test.drop(columns=['Cabin','Name'],axis=1,inplace=True)
```

We have CryoSleep and VIP features which has values in string as True and False.But system will not understand this string type.so we need to change True to 1 and False to 0. We can do it by changing data type as integer.

```python
# Change true -> 1 and False -> 0 
df_train['CryoSleep']=df_train['CryoSleep'].astype('int')
df_train['VIP']=df_train['VIP'].astype('int')

df_test['CryoSleep']=df_test['CryoSleep'].astype('int')
df_test['VIP']=df_test['VIP'].astype('int')
```

#### Encoding 
Encoding is a technique to convert categorical data to numerical so that machine can understand and train it.Here we have HomePlanet and Destination which have categorical data,showing below :-

```python
# count of all the distinct values in HomePlanet(a categorical feature)
print("train",df_train['HomePlanet'].value_counts())
print("-"*30)
print("test",df_test['HomePlanet'].value_counts())
```

Here we are using one of the encoding technique that is LabelEncoding.It will covert all the categorical values to numerical ones like -
for Homeplanet - There are three categories - Earth, Europe and Mars.so it will convert these three into 0,1 and 2.

```python
# LabelEncoding for train and test data for Homeplanet and Destination Features
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
df_train['HomePlanet']=Le.fit_transform(df_train['HomePlanet'])
df_train['Destination']=Le.fit_transform(df_train['Destination'])

df_test['Destination']=Le.fit_transform(df_test['Destination'])
df_test['HomePlanet']=Le.fit_transform(df_test['HomePlanet'])
```

#### Building Models <br>
First we need to split our df_train dataset into train and test using train_test_split method.<br>
Test_size has set 0.3 it means df_train will be splitted into two parts.In which 70% is training data and 30% is test data.

```python
# splitting the train dataset into train and test dataset using train_test_split method
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df_train,y,random_state=20,test_size=0.3)
```

```python
#RandomForest classifier for training dataset
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
rm_f = RandomForestClassifier(max_depth=7,random_state=0)
rm_f.fit(X_train,y_train)
a = rm_f.predict(X_test)
print(accuracy_score(a,y_test))
```

```python
# Logistic regression classifier for training
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state=0)
log_reg.fit(X_train,y_train)
b = log_reg.predict(X_test)
print(accuracy_score(b,y_test))
```

```python
# DecisionTree classifier for training 
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(random_state=0)
dt_clf.fit(X_train,y_train)
c = dt_clf.predict(X_test)
print(accuracy_score(c,y_test))
```

#### Output Submission <br>
As we have seen above that best accuracy score is for RandomForest Classifier and that is around 80%.so making submission on behlaf on randomforest model.<br>
Assiging results of randomforest model to y_pred and passengerid to id. y_pred and id are the variables which we are going to place in submission file.

```python
y_pred = rm_f.predict(df_test)
id = df_test.PassengerId
```
